# Data Cleaning

## Dataset Distrito

```python
# Ahora averiguaré si el dataset tiene filas duplicadas.
distrito.drop_duplicates(inplace=True)
distrito.shape
```

Como podemos apreciar el dataset aparenta no tener filas duplicadas, ya que contiene la misma forma que en nuestra primera instancia de la función (shape).

```python
# Vamos a crear una copia de nuestro dataset para ver si tiene valores nulos.
distrito_cp = distrito.copy()
```

```python
# Usaremos (dropna) para remover las filas con valores nulos.
distrito_cp.dropna(axis = 0 , inplace=True ) 
# Ahora checaremos con (info).
distrito_cp.info()
```

Ahora si comprobamos con la cantidad de valores de nuestro primer instancia de la función (info) podemos percatarnos de que los valores son los mimos por ende no hay valores duplicados.
